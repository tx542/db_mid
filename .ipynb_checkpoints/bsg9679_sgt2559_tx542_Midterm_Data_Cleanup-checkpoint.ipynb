Usually, when Pandas cannot automatically recognize a column variable as `numeric`, it is because there exist some (row) entries to such a variable that cannot be implicitly converted into numeric data types (int or float). 

Let us define a function below that is able to print out the list of unique values for each column of a Pandas DataFrame, and call the function on each column variable of our `new_salary_df` DataFrame:

```python
def unique_val_per_col(df):
    '''
    This function prints the list of unique values
    for each column of the given Pandas DataFrame
    '''
    for col_var in (df.columns.tolist()):
        
        print("The unique values of the variable named ", end = "")
        print('\033[1;37;40m"{}"'.format(col_var), end = "")
        # having some fun with ANSI escape code
        # to change the format of the output text
        print("\033[0;0;0m are:\n",
              new_salary_df[col_var].unique(), "\n",
              sep = "\n")
    
    return None
```

```python
# call the function (defined above)
# to see the list of unique values for all cols
# of our new_salary_df DataFrame

unique_val_per_col(new_salary_df) 
```

# Section II: Categorical Data

Having fixed the intrinsic problem associated with real-world data (especially those fields that allow any user input without conditional/data-type checks) by now, we have made the columns of data that have numeric significance (such as years of experience, salary, or the number of vacation days) numeric—as series of floats.

Now, referring back to <a href = "https://docs.google.com/forms/d/e/1FAIpQLSdPDpjEN98tazCLOQ7xxgK84DZeanC8wI_akPyKOeW3HwBhuA/viewform">the original survey</a>, and in consideration of consistency with the original researchers' intentions, we will also convert some column variables to categorical ones, where applicable.

In particular, we will focus on those variables that correspond to questions where the answers are pre-filled choices (and sometimes with an "Other" option that allows for any user input). We will define a function `make_categorical` below, specifically for this purpose:

```python
def make_categorical(df, col_label, cate_ls, exist_other = True, order = True):
    '''
    A custom-made function to turn the column with given label into a 
    Pandas categorical column. 
    
    The parameter 'exist_other' indicates whether a "Other"
    category (everything other than the pre-filled categories) exists.
    
    The parameter 'order' indicates whether the categorical column
    shall be ordered or not. 
    '''
    
    if col_label not in df.columns.tolist():
        # checks if given column label is really a column in the given df
        print("Error: given label not in given DataFrame")
        return None
    
    if exist_other == True:
        # locate all entries to given column label 
        # where the value is NOT in the category list
        # using bitwise boolean operator ~ to negate the Boolean expression
        # Replace those values with "Other"
        df[col_label].loc[~ df[col_label].isin(cate_ls)] = "Other"
        cate_ls = cate_ls + ["Other"]
        
    df[col_label] = pd.Categorical(df[col_label],\
                                     categories = cate_ls,\
                                     ordered = order)
    # create categorical data and assign it back (to replace)
    # the given column label in the given df
    
    print(df[col_label].value_counts())
    # printing the resulting value count
    
    return None
```
