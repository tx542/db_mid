### 2. Histograms: Numeric Data and Distribution

In addition to listing numeric statistics in tables, some nice graphic illustrations of the distribution of the column variables would be more visually appealing and direct in their storytelling. And in the case of numeric variables, we would be able to construct histograms that demonstrate the distribution densities of such variables throughout various intervals (which can differ depending on the range of values that the variable can take on). 

Hence, let us start by taking a look at the minimum and maximum values for each of the numeric variables of our DataFrame:

```python
new_salary_df.agg(["min", "max"])
```

We see some problems with extreme outlier values—especially for the variables related to salaries and bonuses (this year or last). In our operations below, we will drop the significant outliers, so that the distribution or scale of our histograms would *not* be driven by those very few (at most three) outliers alone.

```python
# set the plotnine figure size
plotnine.options.figure_size = (16, 12)
```

```python
def plot_hist(var_name, custom_colors = None):
    '''
    A function that can make a histogram for the given column label 
    (corresponding to a column variable) of our new_salary_df DataFrame
    '''
    
    graph_data = new_salary_df.dropna(subset = [var_name])
    
    # manually removing the extreme outliers so our graphs are not driven by them alone
    if var_name == "sal":
        graph_data["sal"] = graph_data["sal"].replace(1.000000e+11, np.NaN)
        graph_data["sal"] = graph_data["sal"].replace(99999999999.0, np.NaN)
        graph_data["sal"] = graph_data["sal"].replace(500000000.0, np.NaN)
        
    elif var_name == "sal_last":
        graph_data["sal_last"] = graph_data["sal_last"].replace(500000000.0, np.NaN)
        
    elif var_name == "bonus":
        graph_data["bonus"] = graph_data["bonus"].replace(5.000000e+09, np.NaN)
        graph_data["bonus"] = graph_data["bonus"].replace(1150000.0, np.NaN)
        
    elif var_name == "bonus_last":
        graph_data["bonus_last"] = graph_data["bonus_last"].replace(50000000.0, np.NaN)
        
    elif var_name == "vac_days":
        graph_data["vac_days"] = graph_data["vac_days"].replace(365, np.NaN)
        
    # drop NaN if we just added some in the replacements above (where applicable)
    graph_data = graph_data.dropna(subset = [var_name])
    
    # get the value range for the plot of the given column variable
    var_range = tuple(graph_data.agg(["min", "max"])[var_name])
    print("Value range of the histogram for {} is:".format(var_name), var_range)
    
    # manually renaming var labels if too long/where applicable:
    var_x_label = var_to_labels[var_name]
    
    if var_name == "money_sup":
        var_x_label = "Additional monetary support to Work From Home in 2020 in EUR"
    elif var_name == "sal":
        var_x_label = "Annual brutto salary (without bonus and stocks) in 2020 in EUR"
    elif var_name == "sal_last":
        var_x_label = "Annual brutto salary (without bonus and stocks) in 2019 in EUR"
    elif var_name == "bonus":
        var_x_label = "Annual bonus + stocks in 2020 in EUR"
    elif var_name == "bonus_last":
        var_x_label = "Annual bonus + stocks in 2019 in EUR"
    elif var_name == "short_work":
        var_x_label = "Hours of imposed shorter working week (Kurzarbeit)"
    
    obj = (ggplot(graph_data, aes(x = var_name)) 
           + geom_histogram(aes(y = "stat(density)"), alpha = 0.5, fill = "green")
           + geom_density(color = "red", alpha = 0.5)
           + ggtitle('Histogram for the Density Distribution of "{}" with KDE Line'\
                     .format(var_x_label))
           + ylab("Density")
           + xlab(var_x_label)
           + theme_bw())
    
    # save the plot in a local file, named after the given col var label
    obj.save("histograms/hist_{name}.png".format(name = var_name))
    
    return None
```

Let us proceed to make and save the histogram (density distribution) plot for each of our numeric variables:

```python
for num_var in num_var_ls:
    plot_hist(num_var)
```

### 3. Pie Charts: Categorical Data Components and Percentages
Some nice pie charts below, about the components that make up each categorical data and their respective percentages

Given that our DataFrame has, in addition to the numeric variables, categorical variables as well, we will proceed to conduct some very similar descriptive illustrations for our categorical variables—using pie charts, as they have a finite number of possible categories that can be associated with a corresponding percentage value.

We start the experiment with a rudimentary pie chart using the `.plot()` method for a Pandas DataFrame (as we have learned that the pie chart plotting method has not yet been implemented for `plotnine` for Python), using the categorical variable `pos` (Position) as an example:

```python
new_salary_df.groupby("pos").size().plot(kind = "pie", autopct = "%.2f");
```

That can be a start... But we can improve the aesthetics and make the graph appear more visually appealing, using `matplotlib`. Let us define a function that can make a pie chart for a given column label (corresponding to a column variable) of our `new_salary_df` DataFrame:

```python
def plot_pie(var_name, custom_colors = None):
    '''
    A function that can make a pie chart for the given column label 
    (corresponding to a column variable) of our new_salary_df DataFrame
    '''
    plt.clf() # clear any previous figure window
    plt.figure(figsize = (8, 6), dpi = 100) # set figure size
    plt.pie(new_salary_df[var_name].value_counts(), 
            labels = new_salary_df[var_name].value_counts().index,
            autopct = '%.2f%%', # customized colors below:
            colors = ["red", "blue", "orange", "green", "lightskyblue", "gold"],
            wedgeprops = {"alpha": 0.4}) # increase transparency for visual effects
    plt.title('Pie Chart of "{var}"'.format(var = var_to_labels[var_name]))
    plt.axis('equal')
    return plt
```

Let us call the function defined above to make a more visually appealing pie chart for the same column variable `pos` (Position):

### 4. Programming Languages: Order of Preference ("Main" and "Other")

Moving on to another topic of great interest and pertinence to us: given that all the participants to this survey work in the IT industry, what **programming languages** would be the **most popular** in terms of usage (ranked by the total number of participants reporting their usage of that programming language)?

From our DataFrame, we have the two "object" variables `tech_lang` and `tech_lang_otr` that denote "the **Main** Technology/Programming Languages" and "the **Other** Technology/Programming Languages" respectively. 

Nonetheless, the variable `tech_lang` is entirely based on user input. And through our examination of the data, there exists a variety of spellings for the same language—sometimes not just varying lower/upper cases, but even abbreviations (e.g. "JS" for "JavaScript") and spellings in a foreign language (e.g. "Офмф" that is the Russian keyboard output when you type the letters "J", "A", "V", "A" in order). 

Thus, we have constructed a dictionary that maps all those varying spellings to the actual name and spelling convention of the programming language. 

See the function definition and the aforementioned dictionary in the cells below:

```python
def count_occur(container, user_input = False, kvmap = None):
    '''
    This function intakes a list-like object (container), assuming that each item of
    that container is a string with multiple substrings concatenated by a comma.
    It will then count the number of occurence of each of such substrings
    '''
    result_dict = {}
    for item in container:
        # split by comma and strip white spaces
        substr_ls = [substring.strip() for substring in item.split(",")]
        
        for entry in substr_ls:
            if (user_input == False) or ((user_input == True) and (entry.lower() not in kvmap)):
                # if the entries are prefilled (no spelling variations)
                # OR 
                # if the user-inputs are not frequently occuring languages (add unique entries)
                if entry in result_dict.keys():
                    # if entry already present: update count by 1
                    result_dict[entry] += 1
                else:
                    # if not present: construct entry, make count 1 (1st occurence)
                    result_dict[entry] = 1
            else:
                if kvmap[entry.lower()] in result_dict.keys():
                    # if the value corresponding to entry already present:
                    # update count by 1
                    result_dict[kvmap[entry.lower()]] += 1
                else:
                    # if not present: construct the value corresponding to entry
                    # make count 1 (1st occurence)
                    result_dict[kvmap[entry.lower()]] = 1
                
                
    return {key: value for key, value in sorted(result_dict.items(),
                                                key = lambda item: item[1], reverse = True)}
```

```python
# here is the entry of the participant 
# who responded in Russian to the tech_lang input
new_salary_df[new_salary_df["tech_lang"] == "Офмф"]
```

```python
# just checking whether the .lower() string method works for Russian as well...
# because there were entries in Russian letters
"Офмф".lower()
# okay, kudos to Python! truly a global spirit
```

```python
# The dictionary, as mentioned above:
# maps the lower-case or alternative spellings of programming languages
# to their actual name/spelling convention

usr_kvmap = {"python": "Python", 
             "java": "Java", 
             "офмф": "Java",
             "php": "PHP", 
             "javascript": "JavaScript", 
             "js": "JavaScript",
             "c++": "C++",
             "c#": "C#",
             "swift": "Swift",
             "kotlin": "Kotlin",
             "scala": "Scala",
             "go": "Go",
             "typescript": "TypeScript",
             "sql": "SQL",
             "ruby": "Ruby",
             ".net": ".NET",
             "c": "C",
             "ios": "iOS",
             "kubernetes": "Kubernetes",
             "r": "R",
             "aws": "AWS",
             "android": "Android",
             "angular": "Angular",
             "react": "React"
            }
```

Now, we are prepared to derive the top 20 main technology/programming languages:

```python
top_20_main_lang = list(count_occur(new_salary_df["tech_lang"].dropna(),
                               user_input = True, kvmap = usr_kvmap).items())[:20]

# The top 20 "main technology / programming language"
print("The top 20 main technology/programming languages are:\n")
idx = 1
for item in top_20_main_lang:
    print("No.{} - Language: {}, Count: {}\n".format(str(idx), item[0], item[1]))
    idx += 1
```

And we can make an ordered (by the number of counts, descending order) bar graph for this data:

```python
# make our dictionary into a Pandas DataFrame and reset index
top_20_df = pd.DataFrame.from_dict(dict(top_20_main_lang),
                                   orient = "index", columns = ["Count"])
top_20_df.reset_index(inplace = True)

# make the order of precedence (categorical), descending:
order_by_freq = [item[0] for item in top_20_main_lang][::-1]

# make it ordered categorical, so that when we plot the bar graph,
# the bar lengths will be in order
top_20_df["index"] = pd.Categorical(top_20_df["index"],
                                  categories = order_by_freq,
                                  ordered = True)
```

```python
# set the plotnine figure size
plotnine.options.figure_size = (8, 6)

(ggplot(data = top_20_df, mapping = aes(x = "index", y = "Count"))
 + geom_bar(stat = "identity", size = 20, fill = "green", alpha = 0.5)
 + coord_flip()
 + ggtitle("The Top-20 Main Technology/Programming Languages and Counts")
 + xlab("Programming Language")
 + ylab("Count")
 + theme_bw()
).draw();
```

We can do the exact same thing for the top 20 **other** technology/programming languages:

```python
top_20_otr_lang = list(count_occur(new_salary_df["tech_lang_otr"].dropna()).items())[:20]

# The top 20 "other technologies/programming languages" (pre-filled values)
print("The top 20 other technology/programming languages are:\n")
idx = 1
for item in top_20_otr_lang:
    print("No.{} - Language: {}, Count: {}\n".format(str(idx), item[0], item[1]))
    idx += 1
```

```python
# make our dictionary into a Pandas DataFrame and reset index
otr_top_20_df = pd.DataFrame.from_dict(dict(top_20_otr_lang),
                                   orient = "index", columns = ["Count"])
otr_top_20_df.reset_index(inplace = True)

# make the order of precedence (categorical), descending:
otr_order_by_freq = [item[0] for item in top_20_otr_lang][::-1]

# make it ordered categorical, so that when we plot the bar graph,
# the bar lengths will be in order
otr_top_20_df["index"] = pd.Categorical(otr_top_20_df["index"],
                                  categories = otr_order_by_freq,
                                  ordered = True)
```

```python
(ggplot(data = otr_top_20_df, mapping = aes(x = "index", y = "Count"))
 + geom_bar(stat = "identity", size = 20, fill = "green", alpha = 0.5)
 + coord_flip()
 + ggtitle("The Top-20 Other Technology/Programming Languages and Counts")
 + xlab("Programming Language")
 + ylab("Count")
 + theme_bw()
).draw();
```

So, based on the set of participants to this survey regarding those who work in the IT industry in Europe in 2020, the programming language `Python` ranks as:

However, whilst the median separates the higher distribution from the lower in half, how well does the mean really represent the salary of everyone, considering the usually expected high skewness in the distribution of socioeconomic indicators such as salary?

Let us calculate some percentages below:

```python
def calc_sal_percentage(obj1_lab, obj2_lab, this_year = True):
    
    '''
    This function calculates the percent of obj1_lab who earn more than
    the average (mean) salary of obj2_lab for this or last year (depending
    on whether "this_year" argument is True or not)
    '''
    
    if (obj1_lab not in new_salary_df["gender"].value_counts().index.tolist()) or\
    (obj2_lab not in new_salary_df["gender"].value_counts().index.tolist()):
        # verify valid labels (whether an existing gender category or not)
        print("Invalid labels")
        return None
    
    # checks which year (verify Boolean value)
    if this_year == True:
        sal_label = "sal"
        yr_label = "this year (2020)"
    elif this_year == False:
        sal_label = "sal_last"
        yr_label = "last year (2019)"
    else:
        print("this_year argument not of a valid Boolean value")
        return None
    
    # calculate total number of obj1_lab
    total_obj1 = new_salary_df["gender"].value_counts()[obj1_lab]

    # calculate mean obj2_lab salary in given year
    obj2_sal_mean = new_salary_df.groupby("gender")[sal_label].mean()[obj2_lab]

    # calculate the number of obj1_lab who earn above that mean
    num_obj1_above_mean = new_salary_df[new_salary_df[sal_label] > obj2_sal_mean].\
    groupby("gender").size()[obj1_lab]

    # calculate the percentage
    percent_obj1_above_obj2 = "{percent:.2%}"\
    .format(percent = (num_obj1_above_mean / total_obj1))

    print('The percentage of of {obj1} who earn more than the {obj2} average (mean) salary {yr} is:'.format(
    obj1 = obj1_lab, obj2 = obj2_lab, yr = yr_label), percent_obj1_above_obj2)
    
    return None
    
```

# Section IV: Some Brief Predictive Statistics & Visualization

### 1. OLS Linear Regression Models

<hr>

**Note to self: things to consider when experimenting with linear regression models**
- Which dependent/explained variable ($Y_i$) do I want to explain (e.g. salary, years of experience)
- Which independent/explanatory variables ($X_i$) should I use to explain that dependent/explained variable ($Y_i$) (use insights from some economic or social science theories)? 
- Which model (univariate or multivariate; likely the latter) has the best explaining power (measured by $R^2$ or $R_{adj}^2$)?
- What about the predicted coefficient ($\hat{\beta_i}$) for the independent/explanatory variable $X_i$ and the corresponding confidence interval? Can we say anything certain (statistically significant) about the effect of $X_i$ on our dependent/explained variable $Y_i$?

<hr>

### 1.1: Univariate OLS Models

As also mentioned in the subsection above, in this section, we will experiment with OLS linear regression models—which aim to minimize the sum of distances ("residuals squared") between the actual $(x, y)$ pairs/points and the predicted values based on the OLS estimations—to try to explain the variations in one variable (the dependent/explained variable $Y_i$) using one or multiple other variables (the independent/explanatory variable $X_i$).

To start with, let us experiment with some univariate OLS models, where we will try to explain the variations in the variable `sal` with `yr_exp`:

```python
smf.ols('sal ~ yr_exp', new_salary_df).fit().summary()
```

As demonstrated in the regression results table above, we are confident at the 1% significance level to reject the null hypothesis that the total years of experience (`yr_exp`) has no effect on the yearly brutto salary in 2020 (`sal`). Based on the OLS estimation, it is estimated that one more year of experience is associated with **a 1892.8472 euros increase** in the yearly brutto salary in 2020.

Nonetheless, the $R^2$ value of 0.092 means that only **9.2%** of all the variations in the variable `yr_exp` can be explained by this univariate model. That doesn't seem to be quite a sufficient explanation—and we should certainly experiment with other models.

So, what happens when we try to explain the log-transformed `ln_sal` with the same explanatory `yr_exp`?

```python
smf.ols('ln_sal ~ yr_exp', new_salary_df).fit().summary()
```

```python
# given that only our y variable is log-transformed,
# we exponent the estimated coefficient for x, subtract one from it, and multiply by 100
# to interpret the percentage change in y associated with one unit of change in x

(np.exp(smf.ols('ln_sal ~ yr_exp', new_salary_df).fit().params["yr_exp"]) - 1) * 100
```

For this new univariate model, the estimated coefficients are still statistically significant (different from zero) at the 1% significance level. Given that our dependent variable has been log-transformed, our interpretation of the estimated results would be that one more year of experience is associated with **a 2.7387% increase** in the yearly brutto salary in 2020.

Meanwhile, the $R^2$ value has increased to **0.16**: when the salary is log-transformed (so that the skewness is decreased), more of its variations can be explained by the same `yr_exp` variable. 

Still, we might also be interested in seeing the effect of other dependent variables on salary. As we add more dependent variables into the OLS regression model and make it multivariate, we could also potentially avoid omitted variable bias (omitting important explanatory variables, so that the estimated coefficients are biased) and improve the explanatory power of our model.

### 1.2: Multivariate OLS Models

Let us start our experimentation with multivariate OLS models by trying to explain `sal` with both `yr_exp` and the categorical variable `gender`:

```python
smf.ols('sal ~ yr_exp+gender', new_salary_df).fit().summary()
```

Again, all the estimated coefficients are statistically significant at the 1% significance level. Given that our model is now multivariate, when interpreting the economic significance of coefficients, we need to use the assumption of "ceteris paribus" (all other conditions being the same) in order to interpret the partial effect of the change in one independent variable on the dependent variable. 

So, ceteris paribus, on average, it is estimated that:
- One more year of experience is associated with **a 1793.8132 euros increase** in the yearly brutto salary in 2020
- Females earn **9810.2795 euros less** than Males
- Those who identify as diverse in gender earn **85390 euros more** than Males (note, however, that the number of observations is limited for the diverse group, so that inferences regarding the population would not be conclusive)

And **11.4** percent of the variations in salary can be explained by this multivariate model.

What if we add another categorical variable, `senior` that denotes the level of seniority, into the existing multivariate model? Theoretically, the level of seniority can also explain the variations in salary:

```python
smf.ols('sal ~ yr_exp+gender+senior', new_salary_df).fit().summary()
```

Based on the regression results table above, we can see that all estimated coefficients are statistically significant at the 1% significance level, except for that of `senior[T.Middle]`. We can make the following inference:

Ceteris paribus, on average, it is estimated that:
- One more year of experience is associated with **a 952.6470 euros increase** in the yearly brutto salary in 2020
- Females earn **6428.3568 euros less** than Males
- Those who identify as diverse in gender earn **72550 euros more** than Males (note, however, that the number of observations is limited for the diverse group, so that inferences regarding the population would not be conclusive)
- Those of Middle seniority level earn **no significantly different salary** than those of Junior seniority level
- Those of Senior seniority level earn **16960 euros more** than those of Junior seniority level
- Those of Lead seniority level earn **28080 euros more** than those of Junior seniority level
- Those of Head seniority level earn **38130 euros more** than those of Junior seniority level
- Those of Other seniority level earn **23460 euros more** than those of Junior seniority level

And **17.5** percent of the variations in salary can be explained by this multivariate model.

Well, it should indeed come as no surprise that, once the seniority level is taken into account (the higher the seniority level, the higher the estimated salary, ceteris paribus, on average), the partial effects of both the number of years of experience and gender category have decreased in magnitude. And the explanatory power of the model has improved as a result of our decision to add in this categorical variable that controls for seniority level. 

### 1.3: Multivariate OLS Models with An Autoregression Component

To continue, a common practice when dealing with variables that involve different time periods in statistics would be adding an autoregressive component to the explanatory model—so that the value of a variable in a past time period ($t-1$) can be used to explain the value of that same variable in the present time period ($t$).

One can quickly notice that in our DataFrame, both `sal` that measures the annual brutto salary for 2020 (time period $t$) and `sal_last` that measures the annual brutto salary for 2019 (time period $t-1$) are present. Let us continue with our experiments by adding the autoregressive variable `sal_last` as an explanatory component to our existing multivariate model, in an attempt to better explain `sal` in the present time. 

```python
smf.ols('sal ~ yr_exp+gender+senior+sal_last', new_salary_df).fit().summary()
```

As illustrated in the regression results table above, one can immediately recognize that the estimated coefficients for several categories of different categorical variables have become **not statistically significant at the 5% significance level**, which suggests the lack of sufficient evidence to reject the null hypothesis (at the 5% significance level) that those categories have no statistically significant difference than the category omitted in the regression table for comparison. 

In other words, as the estimated coefficients for `gender[T.Female]`, `senior[T.Middle]`, and `senior[T.Other]` are all **not statistically significant at the 5% significance level**, ceteris paribus, on average, it is estimated that:
- Females earn **no significantly different salary** than Males
- Those of Middle seniority level earn **no significantly different salary** than those of Junior seniority level
- Those of Other seniority level earn **no significantly different salary** than those of Junior seniority level

On the other hand, ceteris paribus, on average, it is estimated that:
- One more year of experience is associated with **a 796.3139 euros increase** in the yearly brutto salary in 2020
- **One more euro of yearly brutto salary in 2019** is associated with **a 0.3393 euros increase in the yearly brutto salary in 2020**
- Those who identify as diverse in gender earn **64070 euros more** than Males (note, however, that the number of observations is limited for the diverse group, so that inferences regarding the population would not be conclusive)
- Those of Senior seniority level earn **12650 euros more** than those of Junior seniority level
- Those of Lead seniority level earn **16930 euros more** than those of Junior seniority level
- Those of Head seniority level earn **28100 euros more** than those of Junior seniority level

And **23.4 percent** of the variations in salary can be explained by this multivariate model.

Nonetheless, note that the above model estimation is only contingent upon the presence of an autoregressive component in the multivariate OLS model (`sal_last` that controls for the annual brutto salary for 2019 (time period $t-1$)).

Thus, the "no statistically significant difference in salary" between Females and Males, as well as among Middle, Junior, and Other seniority level categories, is only the case **after the autoregressive component for time period $t-1$ salary has been accounted for in the model**. Similarly, the estimated coefficients for all other statistically significant explanatory variables/categories have decreased in magnitude as a result of this same process.

### 1.4: Multivariate OLS Models with Log-Transformed Dependent Variable

Having experimented with several multivariate OLS models to explain `sal`, let us also attempt to explain `ln_sal`, the log transformation of brutto salary in 2020, using `yr_exp` and `gender`:

```python
log_y_mod_1 = smf.ols('ln_sal ~ yr_exp+gender', new_salary_df).fit()
log_y_mod_1.summary()
```

As only our dependent variable `ln_sal` is log-transformed, we need to conduct some additional operations with the estimated coefficients for the non-log-transformed independent variables, as defined in the function below:

```python
def log_y_interpret(x_coeff):

    # given that only our y variable is log-transformed,
    # we exponent the estimated coefficient for x, subtract one from it, and multiply by 100
    # to interpret the percentage change in y associated with one unit of change in x
    
    return "{:.4f}%".format((np.exp(x_coeff) - 1) * 100)
```

```python
log_y_interpret(log_y_mod_1.params["yr_exp"])
```

Ceteris paribus, on average, it is estimated that one more year of experience is associated with **a 2.6056% increase** in the yearly brutto salary in 2020.

```python
log_y_interpret(log_y_mod_1.params["gender[T.Female]"])
```

Ceteris paribus, on average, it is estimated that Females earn **11.6403% less** in the yearly brutto salary in 2020 than Males.

```python
log_y_interpret(log_y_mod_1.params["gender[T.Diverse]"])
```

Ceteris paribus, on average, it is estimated that those who identify as diverse in their gender earn **156.2741% more** in the yearly brutto salary in 2020 than Males. (Note, however, that the number of observations is limited for the diverse group, so that inferences regarding the population would not be conclusive)

And **18.7** percent of the variations in log-transformed salary can be explained by this multivariate model.

Let us also add an autoregressive component, `ln_sal_last` (the log transformation of brutto salary in 2019) to the model:

```python
log_y_mod_2 = smf.ols('ln_sal ~ yr_exp+gender+ln_sal_last', new_salary_df).fit()
log_y_mod_2.summary()
```

After adding `ln_sal_last` into the model, some very interesting things happened: 
- Once the autoregressive component of `ln_sal` is accounted for, the model estimations suggest that ceteris paribus, on average, Females earn no significantly different salary than Males (at the 5% significance level). 
- Also, as much as **75.1** percent of the variations in log-transformed salary can be explained by this multivariate model. 

Does this necessarily mean that we have found the perfect model to explain the variations in salary in the IT industry in Europe in 2020?

Well, not necessarily!

If you still recall, the correlation coefficient between `ln_sal` and `ln_sal_last` is 0.86, which means that these two variables are very strongly, positively correlated. This suggests that even `ln_sal_last` alone could explain the variations in `ln_sal`, as shown below:

```python
smf.ols('ln_sal ~ ln_sal_last', new_salary_df).fit().summary()
```

This very simple, univariate, autoregression (AR) model suggests that the salary in 2020 is highly contingent ($R^2 = 0.745$) upon the salary in 2019 (time period $t-1$, with an estimated coefficient of 0.8058), plus some errors that resemble white noise. 

However, we are not able to attempt to fit other AR models that might involve two or more past time periods (e.g. $t-1$ salary and $t-2$ salary), as we do not have the relevant data. Furthermore, it is theoretically unlikely that the variations in salary occurred due to a stochastic process, free of other socioeconomic or political factors. 

Additionally, though a person's salary might be correlated with what they earned in the past, we would still need to determine the contributing factors to their starting salary (initial value, at $t_0$).

Therefore, even though in a statistical sense, we could explain `ln_sal` using even only `ln_sal_last`, our economic intuitions might meanwhile suggest that such a relationship is of a "correlation, and not necessarily causation" nature and demand that we look for additional, confounding explanatory variables. 

### 1.5: Explanatory Power vs. Direction of Causality
Continuing the previously mentioned, classic discussion of "correlation vs. causation":

When choosing from a range of possible models that explain the same dependent variable, some of the things that economists and statisticians would focus on as selection criteria could be:
- The statistical significance of estimated coefficients
- The explanatory power of the model ($R^2$, or $R_{adj}^2$ if comparing multiple models with different numbers of explanatory variables)
- Parsimony: if both criterias above are similar, models with fewer explanatory variables are preferred for simplicity and generalizability 

With those principles in mind, let us move on to our final topic of discussions regarding OLS models. 

If you still recall, we were also interested in explaining variations in the variable `yr_exp` (the total years of experience) in earlier sections. 

After quite some experimentations, and following the criteria listed above, our group has determined the following model to be the optimal choice for explaining `yr_exp`:

$$yr\_exp = \beta_0 + \beta_1 age + \beta_2 senior + \epsilon_i$$

And the model estimation would be:

$$\hat{yr\_exp} = \hat{\beta_0} + \hat{\beta_1} age + \hat{\beta_2} senior$$


```python
smf.ols('yr_exp ~ age+senior', new_salary_df).fit().summary()
```

As illustrated in the regression results table above, the $R^2$ value suggests that as much as **61.2%** of all variations in `yr_exp` can be explained by this model. Taking a further look at the estimated coefficients, we can further notice that all of them are statistically significant at the 1% significance level. 

Thus, ceteris paribus, on average, it is estimated that those of the **Middle, Senior, Lead, Head, and Other** seniority levels have, respectively, **2.3649, 4.6295, 5.9123, 6.1859, and 3.7799 more total years of experience** than those of the Junior seniority level; and an increase by one year in age is associated with **a 0.5881 year of increase** in the total years of experience. 

Nonetheless, aside from the strong correlation relationships among the independent and dependent variables, and the explanatory power of this model, before necessarily establishing a relationship of causality, one of the first things we should think about would be the direction of causality.

Intuitively, we know that it is likely that someone of a higher seniority level has had a greater number of years of experience, and can potentially infer that the latter has contributed to a greater likelihood in being promoted, thus leading to the former (`yr_exp` causes `senior` categories). In the meantime, we could potentially also argue that a higher seniority level—associated with a distinguished position of socioeconomic power—would causally incentivize someone to spend a longer time in the same industry (instead of, for example, entering another industry as a novice). And this could provide the theoretical foundation for the `senior`-causing-`yr_exp` direction of causation as well. 

Nonetheless, as the dependent variable $Y_i$ cannot be a categorical variable in OLS regression models, we would not be able to experiment with the alternative direction of causality (`yr_exp`-causing-`senior`). Hence, we cannot definitively rule out all other possible directions of causality and confounding variables, and conclude with certainty that `senior` can only unidirectionally cause the variations in `yr_exp`. 

With these analyses in mind, our model estimations can only go as far as indications of correlation (some units of change in $X_i$ being associated with some units of change in $Y_i$), and not indicate definitive relationships of causation. 

### 2. KNN and Random Forests
—Using the K-Nearest Neighbors and Random Forests to model and predict the trend in the relationship between log-transformed salary and years of experience. 

And as the final section of our analyses for this midterm project, we will proceed with some brief experiments with the KNN and Random Forests libraries/algorithms, which we learned only very recently, to derive some predictive models. As the Professor has emphasized that this midterm project shall focus mostly on descriptive analysis, we will utilize only the visual (graphic) appearance of the predicted value curve to approximate optimal parameters to the relevant models, rather than using train-test split and cross-validation. 

### 2.1: KNN

```python
# dropping the NaN observations in these columns 
# so that we may use them for some predictive analyses

exp_salary_df = new_salary_df.dropna(subset = ['ln_sal', 'yr_exp'])
```

```python
# from our experiments, 200 neighbors gives a good fitting
# (predicted values form a curve that appears smooth enough [not overfitting]
# and that still captures the overall, nonlinear trends)

sklearn_knn = knn(n_neighbors=200).fit(X = exp_salary_df[['yr_exp']],
                                       y = exp_salary_df['ln_sal'])

sklearn_knn
```

```python
# create the column of predicted KNN values

exp_salary_df['pred_knn'] = sklearn_knn.predict(exp_salary_df[['yr_exp']])
```

```python
# obtain the score of our fitted model
# NOT too much (overfitting), but still quite explanatory

sklearn_knn.score(X=exp_salary_df[['yr_exp']],y=exp_salary_df['ln_sal'])
```

```python
(ggplot(exp_salary_df, aes(x='yr_exp', y='ln_sal'))
 + geom_point(color='green')
 + geom_line(aes(x='yr_exp', y='pred_knn'), color='darkorange', size = 1.5)
 + xlab("Years of Experience")
 + ylab("Log-Transformed Salary")
 + ggtitle("KNN: Log-Transformed Salary vs. Years of Experience")
 + theme_bw())
```

As shown in the K-nearest neighbors (KNN) model fitting graph, the KNN predicted values indeed capture the non-linear relationship between our two variables of concern. Additionally, the fitted model did not significantly fluctuate locally, which suggests that it did not incorporate random noises and errors in the variables. 

### 2.2: Random Forests

```python
# from our experiments, the parameters of 
# 150 total number of models
# and 3 levels of trees in each 
# give a good fitting

sklearn_rf = rf(n_estimators=150,
                max_depth=3).fit(exp_salary_df[['yr_exp']],
                                 exp_salary_df['ln_sal'])
```

```python
# create the column of predicted KNN values

exp_salary_df['pred_rf_2'] = sklearn_rf.predict(exp_salary_df[['yr_exp']])
```

```python
# obtain the score of our fitted model
# NOT too much (overfitting), but still quite explanatory

sklearn_rf.score(exp_salary_df[['yr_exp']],
                                 exp_salary_df['ln_sal'])
```

```python
(ggplot(exp_salary_df, aes(x='yr_exp', y='ln_sal'))+
       geom_point(color='steelblue')+
 geom_line(aes(x='yr_exp', y='pred_rf_2'), color='darkorange' )
  + xlab("Years of Experience")
+ ylab("Log-Transformed Salary")
+ ggtitle("Random Forests: Log-Transformed Salary vs. Years of Experience")
 +theme_bw())
```

Very similar to the KNN fitted model graph, our Random Forests estimations (fitted values) here also demonstrate a nice balance between picking up the non-linear component in the relationship between our variables, and not trying to explain the noises and error terms in the distribution of those two variables that are of a stochastic nature.
